```python
import keras_cv
import matplotlib.pyplot as plt

#model = keras_cv.models.StableDiffusion(img_height=512, img_width=512, jit_compile=True)
model = keras_cv.models.StableDiffusionV2(img_height=512, img_width=512, jit_compile=True)
```

```python
images = model.text_to_image('photo of a standing asian girl wearing nike shirts')
```

```python
def plot_images(images):
  plt.figure(figsize=(20,20))
  for i in range(len(images)):
    ax = plt.subplot(1, len(images), i+1)
    plt.imshow(images[i])
    plt.axis('off')
    plt.tight_layout()
```

```python
plot_images(images)
```
